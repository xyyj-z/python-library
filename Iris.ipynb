````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/Iris.csv')

if 'Id' in df.columns:
    df = df.drop('Id', axis=1)

sns.pairplot(df, hue='Species', palette='husl', markers=["o", "s", "D"])

plt.suptitle("Iris", y=1.02)
plt.show()
````
